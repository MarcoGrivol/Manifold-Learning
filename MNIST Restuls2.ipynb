```python
from manifold_helper import ManifoldHelper
import numpy as np
from collections import OrderedDict
import matplotlib.pyplot as plt
```

```python
methods = ['ISOMAP', 'LLE', 'SE', 'LTSA']
neighbors = [5, 10, 25, 50, 75, 100, 150, 200]
dimensions = [2, 5, 10, 50, 100, 150, 200]
manifold = ManifoldHelper()
```

```python
ari_results_05 = manifold.loadARI(methods, add='id_8')
print("n_percent = 0.05")
manifold.plot_ari_results(ari_results_05, neighbors, dimensions)
```

```python
ari_results_10 = manifold.loadARI(methods, add='id_7')
print("n_percent = 0.1")
manifold.plot_ari_results(ari_results_10, neighbors, dimensions)
```

```python
ari_results_20 = manifold.loadARI(['ISOMAP', 'LLE', 'SE'], add='id_16')
print("n_percent = 0.2")
manifold.plot_ari_results(ari_results_20, neighbors, dimensions)
```

```python
ari_results_diff = OrderedDict()
for m in ['ISOMAP', 'LLE', 'SE']:
    ari_results_diff[m] = ari_results_20[m] - ari_results_05[m]
print("Quanto maior o valor, maior é o benefício de utilzar class_percent maior")
manifold.plot_ari_results(ari_results_diff, neighbors, dimensions)
```
